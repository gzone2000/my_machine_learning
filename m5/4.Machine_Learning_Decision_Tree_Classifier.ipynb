## Split the data and train the Model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(max_depth=4, random_state=0)  # 과대적합을 피하기 위해 가지치기 : max_depth=4 
```

```python
tree.fit(X_train, y_train)
```

```python
pred = tree.predict(X_test)
pred
```

```python
print("Train Set Score: {:.3f}".format(tree.score(X_train, y_train)))
print("Test Set Score: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
print("Feature Importance:\n", tree.feature_importances_)
```

## 결정트리 시각화

```python
export_graphviz(tree, out_file="tree.dot", class_names=["Positive","Negative"], feature_names=cancer.feature_names, impurity=False, filled=True)
```

```python
def plot_feature_importance_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(np.arange(n_features), tree.feature_importances_, align='center')
    plt.yticks(np.arange(n_features),cancer.feature_names)
    plt.xlabel("Feature Importance")
    plt.ylabel('Feature')
    plt.ylim(-1,n_features)
    
plot_feature_importance_cancer(tree)
```
