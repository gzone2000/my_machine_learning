방법1) 이미지 불러오기

```
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```
image = Image.open('./golder-retriever-puppy.jpeg')
image = np.array(image)
image.shape 
```

방법2) 이미지 불러오기

```
import tensorflow as tf
```

```
gfile = tf.io.read_file('./golder-retriever-puppy.jpeg')
image = tf.io.decode_image(gfile)
image.shape   
```

```
plt.title('Dog')
plt.imshow(image)
```
