# Matplotlib : python data visualization

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## 선 그래프

```python
x = np.arange(1,10)
y = x*5
print (x)
print (y)
```

```python
plt.plot(x,y)
plt.show()
```

## 색깔

```python
plt.plot(x,y,color='red')
plt.show()
```

## 기타 스타일

```python
# y = x 그래프를 그리시오. (선 색 : green, 선 굵기 : 0.5, 마커 종류 : circle)
plt.plot(x,y,color='green',marker='o',lw=0.5, linestyle='--')
plt.show()
```

## 레이블과 제목

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('y = 5x')
plt.show()
```

## 축 범위

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('y = 5x')
plt.xlim(0, 10)
plt.ylim(0, 50)
plt.show()
```

## 틱

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('y = 5x')
plt.xticks(np.arange(10)) # tick label
plt.yticks(np.arange(50, step=5))
plt.xlim(0, 10)
plt.ylim(0, 50)
plt.show()
```

## 범례

```python
x = np.linspace(-np.pi, np.pi, 256)
cosine = np.cos(x)
sine = np.sin(x)
plt.plot(x, cosine, ls="--", label="cosine")
plt.plot(x, sine, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
```

## Matplotlib 그림의 구조
- Figure : 그림이 그려지는 캔버스 또는 종이
- Axes : 하나의 plot (한 개 이상)
- Axis : 가로축 or 세로축 (두 개 이상)

![title](https://github.com/gzone2000/repo1/raw/master/image/matplotlib_image.png)

```python
fig = plt.figure()

ax = fig.add_subplot(111)
ax.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) # 위치 (0, 0)
ax1.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

ax2 = fig.add_axes([0.5,0.5,2,2]) # 위치 (0.5, 0.5)
ax2.scatter(np.linspace(0, 2, 10), np.linspace(0, 10, 10))

plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,2)
ax1.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

ax2 = fig.add_subplot(2,2,3)
ax2.scatter(np.linspace(0, 2, 10), np.linspace(0, 10, 10))

plt.show()
```

## How To Change The Size of Figures

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10))

ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])

plt.show()
```

## Options

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10))

ax1.bar([1,2,3],[3,4,5], color='r')
ax1.set_title('vertical bar plot')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend(['axes1'])
ax1.set_ylim([0,10])

ax2.barh([0.5,1,2.5],[0,1,2], color='g')
ax2.set_title('horizontal bar plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend(['axes2'])
ax2.set_xlim([0,3])

plt.show()
fig.savefig('test.png')
```
